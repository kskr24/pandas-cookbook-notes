```python
college=pd.read_csv("https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook/master/data/college.csv",index_col="INSTNM")
college.head()
```

```python
college_ugds_=college.filter(like='UGDS_')
```

```python
college_ugds_.head()
```

```python
college_ugds_.count()
```

```python
college_ugds_.count(axis='index')
```

```python
college_ugds_.count(axis='columns').head()
```

```python
college_ugds_.sum(axis=1)
```

```python
#to get an idea of the distribution of each column, the median method can be used
college_ugds_.median(axis=0)
```

```python
college_ugds_.isnull()\
    .sum(axis=1)\
    .sort_values(ascending=False)\
    .head()
```

```python
college_ugds_=college_ugds_.dropna(how='all')
college_ugds_.isnull().sum()
```

```python
#summary stats for object and categorical columns
college.describe(include=[np.object,pd.Categorical]).T
```

```python
#Reducing memory by changing data types
#1. Reading a few columns
college.index
```

```python
college.head()
```

```python
college=college.drop(['index','level_0'],axis=1)
```

```python
college.head()
```

```python
different_cols=['RELAFFIL','SATMTMID','CURROPER','INSTNM','STABBR']
col2=college[different_cols]
col2=col2.loc[:,:]
```

```python
col2.head()
```

```python
col2.dtypes
```

```python
original_mem=col2.memory_usage(deep=True)
original_mem
```

```python
col2['RELAFFIL']=col2['RELAFFIL'].astype(np.int8)
```

```python
col2.dtypes
```

```python
college[different_cols].memory_usage(deep=True)
```

```python
new_mem=col2.memory_usage(deep=True)
new_mem
```

```python
new_mem/original_mem
```
